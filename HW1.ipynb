### Условие 2:
На складе лежат разные фрукты в разном количестве.
Нужно написать функцию, которая на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе

```python
def count_fruits(fruits: dict) -> float:
    """ Функция на вход принимает любое количество названий фруктов и их количество, а возвращает общее количество фруктов на складе """
    total = 0
    for fruit, quantity in fruits.items():
        total += quantity
    return total


fruits = {
    "яблоки": 40,
    "бананы": 16,
    "апельсины": 25
}
print(count_fruits(fruits))
```

#### 4.2 Найдите выручку компании в зависимости от месяца
Для этого напишите функцию, которая на вход принимает список с датами и список с выручкой, а на выходе словарь, где ключи - это месяцы, а значения - это выручка.
Используйте аннотирование типов.

```python
from datetime import datetime


def monthly_revenue(dates: list, amounts: list) -> dict:
    """ Функция на вход принимает список с датами и список с выручкой, а возвращает словарь, где ключи - это месяцы, а значения - это выручка """
    monthly_revenue = {}
    for date, amount in zip(purchase_dates, purchase_amounts):
        month = datetime.strptime(date, '%Y-%m-%d').strftime('%Y-%m')
        if month in monthly_revenue:
            monthly_revenue[month] += amount
        else:
            monthly_revenue[month] = amount
    return monthly_revenue


purchase_dates = ['2021-09-14', '2021-12-15', '2021-09-08', '2021-12-05',
                  '2021-10-09', '2021-09-30', '2021-12-22', '2021-11-29',
                  '2021-12-24', '2021-11-26', '2021-10-27', '2021-12-18',
                  '2021-11-09', '2021-11-23', '2021-09-27', '2021-10-02',
                  '2021-12-27', '2021-09-20', '2021-12-13', '2021-11-01',
                  '2021-11-09', '2021-12-06', '2021-12-08', '2021-10-09',
                  '2021-10-31', '2021-09-30', '2021-11-09', '2021-12-13',
                  '2021-10-26', '2021-12-09'
                  ]
purchase_amounts = [1270, 8413, 9028, 3703, 5739, 4095, 295, 4944, 5723, 3701, 4471, 651, 7037,
                    4274, 6275, 4988, 6930, 2971, 6592, 2004, 2822, 519, 3406, 2732, 5015, 2008, 316, 6333, 5700, 2887]

print(monthly_revenue(purchase_dates, purchase_amounts))
```
